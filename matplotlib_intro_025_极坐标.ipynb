```python
import numpy as np
import matplotlib.pyplot as plt

r = np.arange(1, 6, 1)
theta = [0, np.pi/2, np.pi,3*np.pi/2,2* np.pi]
ax = plt.subplot(111, projection='polar');
ax.plot(theta, r, color='r', linewidth=3)
ax.grid(True)
```

```python
r = np.empty(9)
r.fill(5)
pi_two = np.pi * 2
theta = [0, pi_two/8, 2*pi_two/8, 3*pi_two/8, 4*pi_two/8, 5*pi_two/8, 6*pi_two/8, 7*pi_two/8, 8*pi_two/8]

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color='r', linewidth=2)
plt.show()
```

```python
n = 5
r = np.empty(n)
r.fill(1)
pi_two = np.pi * 2
theta = [0, pi_two/n, 2*pi_two/n, 3*pi_two/n, 4*pi_two/n]

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color='r', linewidth=2)
plt.show()
```
