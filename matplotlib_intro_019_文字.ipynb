```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-11,12,1)
# y = x*x
y = x**2

plt.plot(x, y);
plt.text(-5, 50, 'function:y=x*x', family='serif', size=20, color='r', style='italic', weight='book')
plt.text(-2.5, 20, 'function:y=x*x', family='fantasy', size=10, color='b', style='oblique', weight='light', bbox=dict(facecolor='r', alpha=0.2))
```

```python
y = x**3
plt.plot(x, y)
plt.text(-5, 500, 'function:y=x**3', family='fantasy', size=15, color='b', style='italic', weight='book', bbox=dict(facecolor='g', alpha=0.4))
```
