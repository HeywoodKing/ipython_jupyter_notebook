```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] # 用于正常显示中文标签
plt.rcParams['axes.unicode_minus']=False   # 用来正常显示负号

x = np.arange(0,10,1)
y = np.random.randn(len(x))

plt.title('bbb')
plt.plot(x,y);
```
