```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

y = np.arange(1, 5)

plt.plot(y, color='g');
plt.plot(y + 1, color='0.5');
plt.plot(y + 2, color='#ff00ff');
plt.plot(y + 3, color=(0.1,0.2,0.3));
plt.plot(y + 4, color='#987654');
```

```python
plt.plot(y, 'o');
plt.plot(y + 1, marker='D');
plt.plot(y + 2, marker='^');
plt.plot(y + 3, marker='p');
plt.plot(y + 4, marker='h');
```

```python
plt.plot(y, '--');
plt.plot(y + 1, '-.');
plt.plot(y + 2, ':');
plt.plot(y + 3, '-');
plt.plot(y + 4, '+');
```

```python
plt.plot(y, 'cx--');
plt.plot(y + 1, 'kp:');
plt.plot(y + 2, 'mo-.');
```
