```python
#  散点图
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

height = np.random.randint(150,200,6)
weight = np.random.randint(40,100,6)

plt.scatter(height,weight);
```

```python
N = 500
x = np.random.randn(N)
y = np.random.randn(N)
plt.scatter(x,y);
```

```python
x = np.random.randn(N)
y = -x + np.random.randn(N) * 0.5
plt.scatter(x,y);
```

```python
open,close = np.loadtxt('000001.csv',delimiter=',',skiprows=1,usecols=(1,4),unpack=True)
change = close - open
yesterday = change[:-1]
today = change[1:]
plt.scatter(yesterday,today,s=50,c='gray',marker='+',alpha='0.5');
```

```python
open,high = np.loadtxt('000001.csv',delimiter=',',skiprows=1,usecols=(1,2),unpack=True)
diff = high - open
yesterday = diff[:-1]
today = diff[1:]
plt.scatter(yesterday,today,s=30,c='red',marker='>',alpha='0.4');
```
