```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(2,20,1)
y1 = x*x
y2 = np.log(x)

# plot方式
plt.plot(x,y1);
plt.twinx()
plt.plot(x,y2,'r');
```

```python
# 面向对象方式
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x,y1);
ax1.set_ylabel('y1')
ax1.set_xlabel('compare y1 and y2');

ax2 = ax1.twinx()
ax2.plot(x,y2, 'r');
ax2.set_ylabel('y2');
```
