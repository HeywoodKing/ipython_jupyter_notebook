```python
#  饼图
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

labels = 'A','B','C','D'
x = np.random.randint(0,100,4)
explode = [0,0.1,0.08,0]
plt.axes(aspect=1)
plt.pie(x=x,labels=labels,autopct='%.0f',explode=explode,shadow=True);
```

```python
#  作业
labels = 'SH','BJ','SZ','GZ'
fracs = [20,10,30,25]
explode = [0.01,0.02,0.03,0.05]
plt.axes(aspect=1)
plt.pie(x=fracs,labels=labels,autopct='%.0f',explode=explode,shadow=True);
```
