```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots()

xy1 = np.array([0.2, 0.2])
circle = mpatches.Circle(xy1, 0.05, color='r')
ax.add_patch(circle)

xy2 = np.array([0.2, 0.8])
rect = mpatches.Rectangle(xy2, 0.2, 0.1, color='b')
ax.add_patch(rect)

xy3 = np.array([0.9,0.2])
polygon = mpatches.RegularPolygon(xy3, 6, 0.1, color='g')
ax.add_patch(polygon)

xy4 = np.array([0.9, 0.8])
ellipse = mpatches.Ellipse(xy4, 0.4, 0.2, color='r')
ax.add_patch(ellipse)

plt.axis('equal')
plt.grid()
# plt.show()
```
