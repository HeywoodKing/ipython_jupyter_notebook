```python
import numpy as np
# import matplotlib as mpl
import matplotlib.pyplot as plt
#设置字体为SimHei显示中文
# plt.rcParams['font.sans-serif'] = 'SimHei'
#设置正常显示字符
# plt.rcParams['axes.unicode_minus'] = False
#设置线条样式
# plt.rcParams['lines.linestyle'] = '-.'
#设置线条宽度
# plt.rcParams['lines.linewidth'] = 3
plt.rcParams['figure.figsize'] = (10.0, 10.0)     # 显示图像的最大范围
# plt.rcParams['image.interpolation'] = 'nearest' # 差值方式，设置 interpolation style
# plt.rcParams['image.cmap'] = 'gray'             # 灰度空间
# plt.rcParams['axes.prop_cycle'] = ['r', 'g', 'b', 'c']# 系列颜色
plt.rcParams['axes.linewidth'] = 1 # 图表边框
plt.rcParams['axes.titlesize'] = 12 # title大小

# print(plt.rcParams.keys())
# plt.style.use('seaborn-pastel')
plt.style.use('ggplot')
# plt.style.use('grayscale')
# print(plt.style.available)


fig, axes = plt.subplots(ncols=2, nrows=2)
# 将4个图的坐标轴作为4个对象
ax1, ax2, ax3, ax4 = axes.ravel()

x, y = np.random.normal(size=(2, 100))
ax1.plot(x, y, 'o');

x = np.arange(0, 10)
y = np.arange(0, 10)
ncolors = len(plt.rcParams['axes.prop_cycle'])
# 颜色自动循环
# print(plt.rcParams['axes.prop_cycle'])

shift = np.linspace(0, 10, ncolors)
for s in shift:
    ax2.plot(x, y+s, '-');


x = np.arange(5)
y1, y2, y3 = np.random.randint(0, 25, size=(3,5))
width = 0.3
# ax3.bar(x, y1, width)
# ax3.bar(x+width, y2, width)
# ax3.bar(x+2*width, y3, width)

# cycler = plt.rcParams['axes.prop_cycle']
# print(cycler)
"""
b ：Blue
g ：Green
r ：Red
c ：Cyan
m： Magenta
y ：Yellow
k ：Black
w ：White
"""
ax3.bar(x, y1, width, color='r')
ax3.bar(x+width, y2, width, color='g')
ax3.bar(x+2*width, y3, width, color='b')

# for color in plt.rcParams['axes.prop_cycle']:
for i, color in enumerate(plt.rcParams['axes.prop_cycle']):
#     print(color)
    xy = np.random.normal(size=20)
    ax4.add_patch(plt.Circle(xy, radius=0.1, color=color['color']))
    
ax4.axis('equal')

plt.show()
```
