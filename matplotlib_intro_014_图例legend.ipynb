```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11, 1)
plt.plot(x, x*2, label='Normal');
plt.plot(x, x*3, label='Fast');
plt.plot(x, x*4, label='Faster');

# plt.legend(loc=4);
plt.legend(loc=0, ncol=2);
```
