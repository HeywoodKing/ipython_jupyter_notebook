```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

# x = np.arange(-10, 10, 1)
# y = x**2 + x*3 + 8

# plt.plot(xy);
# plt.scatter(x, y, color='g', linewidth=9);

x = np.random.randn(200)
y = x + np.random.randn(200) * 0.5

margin_border = 0.1
margin_between = 0.02
width = 0.6
height = 0.2

left_s = margin_border
bottom_s = margin_border
height_s = width
width_s = width

left_x = margin_border
bottom_x = margin_border + width + margin_between
height_x = height
width_x = width

left_y = margin_border + width + margin_between
bottom_y = margin_border
height_y = width
width_y = height

plt.figure(1, figsize=(8, 8))
rect_s = [left_s, bottom_s, width_s, height_s]
rect_x = [left_x, bottom_x, width_x, height_x]
rect_y = [left_y, bottom_y, width_y, height_y]

ax_scatter = plt.axes(rect_s)
ax_his_x = plt.axes(rect_x)
ax_his_y = plt.axes(rect_y)

ax_his_x.set_xticks([])
ax_his_y.set_yticks([])

ax_scatter.scatter(x, y);


bin_width = 0.25
xy_max = np.max([np.max(np.fabs(x)), np.max(np.fabs(y))])
lim = int(xy_max/bin_width + 1) * bin_width
ax_scatter.set_xlim(-lim, lim)
ax_scatter.set_ylim(-lim, lim) 

bins = np.arange(-lim, lim + bin_width, bin_width)

ax_his_x.hist(x, bins=bins, color='g');
ax_his_y.hist(y, bins=bins, orientation='horizontal', color='b');

ax_his_x.set_xlim(ax_scatter.get_xlim())
# ax_his_x.set_ylim(ax_scatter.get_ylim())

# ax_his_y.set_xlim(ax_scatter.get_xlim())
ax_his_y.set_ylim(ax_scatter.get_ylim())

# plt.title('Scatter and Hist', verticalalignment='bottom', loc='left', y=-0.1)
ax_scatter.set_title('Scatter and Hist', y=-0.1, loc='center')

plt.show()
```
